Skripti vattenfallin sivuilta ladattujen hintatiedostojen yhdist√§miseen. Tallenna xlsx tiedostot hinta_data nimiseen kansioon ennen skriptin ajamista.

```python
import pandas as pd
import os
```

```python
files = os.listdir("./hinta_data")
```

```python
df = pd.DataFrame()
df_combined = pd.DataFrame()
for file in files:
    print(file)
    path = "./hinta_data/" + str(file)
    df = pd.read_excel(path, engine='openpyxl')
    df.rename(columns={ df.columns[1]: "Hinta" }, inplace = True)
    df_combined = pd.concat([df_combined, df])
    
df_combined[['Vuosi','Kuukausi', 'Paiva']] = df_combined['Ajanjakso'].str.split('-',expand=True)
df_combined[['Paiva', 'Kellonaika']] = df_combined['Paiva'].str.split(' ',expand=True)
df_combined[['Kellonaika', 'turha1', 'turha2']] = df_combined['Kellonaika'].str.split(':',expand=True)
```

```python
df_combined.drop(['Ajanjakso', 'turha1', 'turha2'], axis=1, inplace=True)
```

```python
df = df_combined
file_name = "porssidata"
df.to_pickle(file_name)  
```
